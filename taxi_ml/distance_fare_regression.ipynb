```python
%matplotlib inline
import numpy as np
import json
import matplotlib.pyplot as plt
import matplotlib as mpl
import csv
x = []
y = []
with open('distance_vs_fare.csv') as f:
    data = csv.reader(f, delimiter=',')
    next(data, None)
    ctr = 0
    for row in data:
        x.append(float(row[1]))
        y.append(float(row[0]))
        if ctr > 10000000:
            break
        ctr += 1
    print(ctr)
```

```python
font = {'weight' : 'bold',
        'size'   : 14}

plt.rc('font', **font)

plt.figure(figsize = (20, 10))
plt.axis([0.0, 0.22, 0, 60])
plt.scatter(x,y, s=80 ** 0.25)
plt.ylabel('Fare Price ($)')
plt.xlabel('Geographical Distance')

plt.show()
```
