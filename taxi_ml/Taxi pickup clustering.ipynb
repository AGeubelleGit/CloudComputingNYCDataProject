```python
%matplotlib inline
import numpy as np
import json
import matplotlib.pyplot as plt
import matplotlib as mpl
import csv
x = []
y = []
colors = []
with open('final3.csv') as f:
    data = csv.reader(f, delimiter=',')
    next(data, None)
    ctr = 0
    for row in data:
        if float(row[0]) > 40.6 and float(row[0]) < 40.9 and float(row[1]) < -73.75 and float(row[1]) > -74.05 :
            x.append(float(row[1]))
            y.append(float(row[0]))
            colors.append(float(row[2])/20.0)
        if ctr > 10000000:
            break
        ctr += 1
    print(ctr)
```

```python
plt.figure(figsize = (20, 10))
# plt.axis([-75, -73, 40, 41])
plt.scatter(x,y,c=colors, s=80 ** 0.25)

plt.show()
```
