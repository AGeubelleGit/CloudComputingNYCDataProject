```python
%matplotlib inline
import numpy as np
import json
import matplotlib.pyplot as plt
import matplotlib as mpl
from descartes import PolygonPatch
mpl.rcParams['agg.path.chunksize'] = 10000
with open('NYCStreetCenterline(CSCL).geojson') as f:
    json_data = json.load(f)
print(json_data.keys())
```

```python
x = []
y = []
for geo in json_data['features']:
    coords = geo['geometry']['coordinates']
    x += [i for i,j in coords[0]]
    y += [j for i,j in coords[0]]
fig = plt.figure(figsize=(20, 10)) 
ax = fig.gca() 
ax.plot(x,y, ',', markersize = 20)
ax.axis('scaled')
plt.show()
```

```python
from matplotlib.colors import LogNorm
import csv
x = []
y = []
with open('yellow_tripdata_2016-01.csv') as f:
    data = csv.reader(f, delimiter=',')
    next(data, None)
    i = 0
    for row in data:
        x.append(float(row[5]))
        y.append(float(row[6]))

heatmap, xedges, yedges = np.histogram2d(x, y, bins=(1024, 1024))
extent = [xedges[0], xedges[-1], yedges[0], yedges[-1]]
 
# Plot heatmap
plt.figure(figsize = (20, 10))
plt.imshow(heatmap, norm=LogNorm(), extent=extent, origin='lower')
plt.colorbar()
plt.show()
```
